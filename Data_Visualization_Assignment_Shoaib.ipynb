## Task

In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

##### Importing relevant libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from sklearn import decomposition
from sklearn import datasets
```

##### Loading Iris Dataset:

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
colours = {'setosa': 'b', 'versicolor': 'r', 'virginica': 'g'}
```

```python
iris['colours'] = iris['species'].map(colours)
```

```python
iris.head()
```

```python
fig  = plt.figure(figsize = (10,8))
ax = fig.gca(projection = '3d')
ax.scatter(iris['petal_length'], iris['petal_width'], iris['sepal_length'],
           zdir='z' , c = iris['colours'],)

ax.legend()
ax.set_xlim(0, 8)
ax.set_ylim(0, 5)
ax.set_zlim(0,9)
ax.set_xlabel('Petal Length')
ax.set_ylabel('Petal Width')
ax.set_zlabel('Sepal Length')
ax.view_init(elev=20., azim=-40)

plt.show()
```
